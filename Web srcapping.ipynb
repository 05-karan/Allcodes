### Request page source from URL

```python
url = "https://www.imdb.com/chart/top/?ref_=nv_mv_250"
```

```python
HEADERS = {'User-Agent': 'Mozilla/5.0 (iPad; CPU OS 12_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Mobile/15E148'}
```

```python
page = requests.get(url, headers=HEADERS)
page
```

```python
#scrap movie names
soup = BeautifulSoup(page.content, 'html.parser')
scraped_movies = soup.find_all('h3', class_='ipc-title__text')
scraped_movies
```

```python
movies = []
for movie in scraped_movies:
    movie = movie.get_text().replace('\n', "")
    movie = movie.strip("")
    movies.append(movie)
movies
```

```python
new_movie = [elem for elem in movies if elem[0].isdigit()]
```

```python
new_movie
best_movies = [elem.translate(str.maketrans('', '', '0123456789.')) for elem in new_movie]
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
scraped_rating = soup.find_all('div', class_='sc-e3e7b191-0 iKUUVe sc-6fa21551-2 kOfhdG cli-ratings-container')
scraped_rating
```

```python
ratings = []
for rating in scraped_rating:
    rating = rating.get_text().replace('\n', "")
    rating = rating.strip("")
    ratings.append(rating[:3])
ratings
```

```python
import pandas as pd
data = pd.DataFrame()
data['Movie Names'] = best_movies
data['Rating'] = ratings
data
```
